```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
    
    def add(self,data):
        NewNode = Node(data)
        NewNode.next = None
        it = self.head
        while(it.next is not None):
            it = it.next
        it.next = NewNode
    
    def printlist(self):
        it = self.head
        while it is not None:
            print(it.data)
            it = it.next
    
    def partition(self,node,val):
        
        head = node
        tail = node
        
        while node!=None:
            Next = node.next
            if(node.data<val):
                node.next = head
                head = node
            else:
                tail.next = node
                tail = node
            node = Next
            tail = node
        tail.next = None
        
        return head
    

list = LinkedList()
list.head = Node(3)
list.add(5)
list.add(8)
list.add(5)
list.add(10)
list.add(2)
list.add(1)
list.printlist()
it = list.head
while it is not None:
    list.partition(it,5)
    it = it.next
print()
list.printlist() 
        
```
