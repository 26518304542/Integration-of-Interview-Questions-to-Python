```python
class Node:
    def __init__(self,data):
        self.data = data
        self.left = self.right = None
        
class Queue:
    def __init__(self):
        self.queue = 2*[0]
        self.last = 0
        
    def append(self,data):
        self.queue += [0]
        self.queue[self.last] = data
        self.last += 1
        
    def pop(self):
        pop_element = self.queue[0]
        i=0
        while i!=self.last:
            self.queue[i] = self.queue[i+1]
            i += 1
        self.queue[self.last] = 0    
        self.last -= 1
        
        return pop_element    
        
        
def pathWay(root,path,k):
    
    if root==None:
        return False
    
    path.append(root.data)
    
    if root.data == k:
        return True
    
    if(root.left!=None and pathWay(root.left,path,k) or root.right!=None and pathWay(root.right,path,k)):
        return True
    
    path.pop()
    return False


def findLCA(root,n1,n2):
    
    path1 = Queue()
    path2 = Queue()
    
    if (not pathWay(root,path1,n1) or not(pathWay(root,path2,n2))):
        return -1
    
    i=0
    while i<path1.last and i<path2.last:
        if path1.queue[i] != path2.queue[i]:
            break
        i += 1
        
    return path1.queue[i-1]
    
    

if __name__ == '__main__':
     
    # Let's create the Binary Tree shown in above diagram
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)
    root.right.left = Node(6)
    root.right.right = Node(7)
     
    print("LCA(4, 5) = %d" % (findLCA(root, 4, 5,)))
    print("LCA(4, 6) = %d" % (findLCA(root, 4, 6)))
    print("LCA(3, 4) = %d" % (findLCA(root, 3, 4)))
    print("LCA(2, 4) = %d" % (findLCA(root, 2, 4)))
        
    
    
    
    
    
    
    
    
    
    
    
    
    


    
```

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.right = self.left = None
        
def findFCA(root,n1,n2):
    
    if root == None:
        return None
    
    if root.data == n1 or root.data == n2:
        return root
    
    left = findFCA(root.left,n1,n2)
    right = findFCA(root.right,n1,n2)
    
    if left and right:
        return root
    
    return  left if left is not None else right

if __name__ == '__main__':
     
    # Let us create a binary tree given in the above example
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)
    root.right.left = Node(6)
    root.right.right = Node(7)
    print("LCA(4, 5) = ", findFCA(root, 4, 5).data)
    print("LCA(4, 6) = ", findFCA(root, 4, 6).data)
    print("LCA(3, 4) = ", findFCA(root, 3, 4).data)
    print("LCA(2, 4) = ", findFCA(root, 2, 4).data)
    
```
