```python
class Stack:
    def __init__(self,capacity):
        self.capacity = capacity
        self.values = self.capacity*[None]
        self.last = 0
    
    def push(self,data):
        if self.isFull()==True:
            self.values += self.capacity*[None]
            self.capacity += self.capacity
        
        self.values[self.last] = data
        self.last += 1
        
    def isFull(self):
        if self.last == self.capacity:
            return True
    
    def peek(self):
        return self.values[self.last-1]
    
    def pop(self):
        if self.isEmpty()==True:
            return
        pop_element = self.values[self.last-1]
        self.values[self.last-1]=None
        self.last-=1
        return pop_element
        
    def isEmpty(self):
        if self.last == 0:
            return True
        
    def printlist(self):
        i=0
        while i!=self.capacity:
            print(self.values[i])
            i+=1
        
    def sort(self):
        s2 = Stack(self.capacity)
        
        while self.isEmpty()!=True:
            temp = self.pop()
            while s2.isEmpty()!=True and s2.peek() > temp:
                self.push(s2.pop())
            s2.push(temp)
            
        while s2.isEmpty()!=True:
            self.push(s2.pop())
            
        
        
s = Stack(3)
s.push(2)
s.push(1)
s.push(4)
s.push(5)
s.push(7)
s.printlist()
s.pop()
print()
s.printlist()
s.sort()
print()
s.printlist()
```
